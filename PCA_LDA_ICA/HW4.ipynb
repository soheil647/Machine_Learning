# Question 9

In forward selection, we start with a null model and then start fitting the model with each individual feature one at a time and select the feature with the maximum score. Now fit a model with two features by trying combinations of the earlier selected feature with all other remaining features. Again select the feature with the maximum score. Now fit a model with three features by trying combinations of two previously selected features with other remaining features. Repeat this process until we have a set of selected features with a p-value of individual features less than the significance level.

In short, the steps for the forward selection technique are as follows :

Choose a significance level (e.g. SL = 0.05 with a 95% confidence).

Fit all possible simple regression models by considering one feature at a time. Total ’n’ models are possible. Select the feature with the lowest p-value.

Fit all possible models with one extra feature added to the previously selected feature(s).

Again, select the feature with a maximum score. we can do this untill all the features are selected. Then we can pick what ever number we want and chose a subset of all the features.

features are all shown below in order of importance

# Step 1
here we group the data set class wise. Meaning each class of our dataset creates an other dataset along with their labels.

```python
def group_data_by_classes(training_data):
    '''
    The data is grouped by classes ie target value of each column
    and also number of classes is stored
    '''
    grouped_data={}
    for index, row in training_data.iterrows():
        if row['label'] in grouped_data:
            grouped_data[row['label']].append(row[:-1])
        else:
            grouped_data[row['label']]=[row[:-1]]
    num_of_classes = len(grouped_data)
    return grouped_data, num_of_classes

display(df)
grouped_data, num_of_classes = group_data_by_classes(df)
print(num_of_classes)
print(len(grouped_data[0]))
```

# Step 2
Here we calculate each class mean and an overall mean along all classes to furthure calculate SB and SW afterwards.

```python
def calculate_means(grouped_data, dim, data_length):
    '''
    This calculates mean of data classwise and overall mean.
    '''
    class_mean={}
    overall_mean=np.array([0. for x in range(dim)])
    for i in grouped_data:
        class_mean[i]=np.array([0. for x in range(dim)])
        for j in grouped_data[i]:
            for k in range(len(j)):
                class_mean[i][k]+=j[k]
                overall_mean[k]+=j[k]
    for i in class_mean:
        for j in range(len(class_mean[i])):
            class_mean[i][j]/=len(grouped_data[i])
    for i in range(len(overall_mean)):
        overall_mean[i]/=len(data_length)

    return class_mean, overall_mean

class_mean, overall_mean = calculate_means(grouped_data, 62, df)         

display(pd.DataFrame(class_mean))
display(pd.DataFrame(overall_mean))
```

# Step 3
Here we calculate both Sw and SB using mean of each class and overall mean of all classes. 

Both SW and SB are 62*62 (d*d) vectors.

```python
def calculate_SB_SW(dim, class_mean, overall_mean, grouped_data):
    '''
    This calculates SB and SW matrices by formula specified in the report
    and return the matrices which are used for data transformation
    '''
    SB = np.zeros((dim, dim))
    for i in class_mean:
        mk_minus_m = np.array([class_mean[i] - overall_mean])
        nk = len(grouped_data[i])
        mk_minus_m_t = mk_minus_m.transpose()
        temp = (mk_minus_m_t * nk).dot(mk_minus_m)
        SB += temp

    SW = np.zeros((dim, dim))
    for i in class_mean:
        mk = np.array(class_mean[i])
        for j in grouped_data[i]:
            xnk = np.array(j)
            xnk_minus_mk = np.array([xnk-mk])
            xnk_minus_mk_t = xnk_minus_mk.transpose()
            SW += xnk_minus_mk_t.dot(xnk_minus_mk)
            
    return SB, SW
                
SB, SW = calculate_SB_SW(62, class_mean, overall_mean, grouped_data)

display(pd.DataFrame(SB))
display(pd.DataFrame(SW))
```

# Step 4
Here we calclate the eigen vectors of Sw-1Sb which are the LDA components.

```python
def calculate_eigen_values(SW, SB, num_dim):
    '''
    This calculates eigenvalues and eigenvectors of the data and number of dimensions
    that data is to be reduced to and calculates vector w that is used for transforming data
    points.
    '''
    mat = np.dot(np.linalg.pinv(SW), SB)
    eigvals, eigvecs = np.linalg.eig(mat)
    eiglist = [(eigvals[i], eigvecs[:, i]) for i in range(len(eigvals))]
    eiglist = sorted(eiglist, key = lambda x : x[0], reverse = True)
    w = np.array([eiglist[i][1] for i in range(num_dim)])
    
    return mat, w, eiglist

seprability_mat, w, eiglist = calculate_eigen_values(SW, SB, 62)

display(pd.DataFrame(seprability_mat))
display(pd.DataFrame(w))
display(pd.DataFrame(eiglist))
```

# Question 14

Here we are going to detect sources using ICA. For this purpose first we create 3 different signals consisting of Sin, Sign, and Sawtooth added by a gaussian noise.

These signals are mixed and our final observation is ploted along with each source signal.

```python
from scipy import signal

# Generate sample data
np.random.seed(0)
n_samples = 2000
time = np.linspace(0, 8, n_samples)

s1 = np.sin(2 * time)  # Signal 1 : sinusoidal signal
s2 = np.sign(np.sin(3 * time))  # Signal 2 : square signal
s3 = signal.sawtooth(2 * np.pi * time)  # Signal 3: saw tooth signal

S = np.c_[s1, s2, s3]
S += 0.2 * np.random.normal(size=S.shape)  # Add noise

plt.figure(figsize=(15,5))
plt.plot(time, s1)
plt.title('Sin')
plt.show()
plt.figure(figsize=(15,5))
plt.plot(time, s2)
plt.title('Square')
plt.show()
plt.figure(figsize=(15,5))
plt.plot(time, s3)
plt.title('Saw Tooth')
plt.show()
plt.figure(figsize=(15,5))
plt.plot(time, S)
plt.title('Mixed')
plt.show()
plt.figure(figsize=(15,5))
plt.plot(time, s1)
plt.plot(time, s2)
plt.plot(time, s3)
plt.show()
```

Here we apply ICA with 3 components (number of signals) to detect each source

```python
from sklearn.decomposition import FastICA

S /= S.std(axis=0)  # Standardize data
# Mix data
A = np.array([[1, 1, 1], [0.5, 2, 1.0], [1.5, 1.0, 2.0]])  # Mixing matrix
X = np.dot(S, A.T)  # Generate observations

# Compute ICA
ica = FastICA(n_components=3)
S_ = ica.fit_transform(X)  # Reconstruct signals
A_ = ica.mixing_  # Get estimated mixing matrix

# We can `prove` that the ICA model applies by reverting the unmixing.
assert np.allclose(X, np.dot(S_, A_.T) + ica.mean_)
```

After applying ICA we can see our observation and true sources and the sources which were found by ICA

It is seen calculated sourses are almost the same as the true sources.

```python
# Plot results
plt.figure()

models = [X, S, S_]
names = ['Observations (mixed signal)',
         'True Sources',
         'ICA recovered signals']

colors = ['red', 'steelblue', 'orange']

fig = plt.figure(figsize=(10,10))
for ii, (model, name) in enumerate(zip(models, names), 1):
    fig.add_subplot(3, 1, ii)
    plt.title(name)
    for sig, color in zip(model.T, colors):
        plt.plot(sig, color=color)

plt.tight_layout()
plt.show()
```
