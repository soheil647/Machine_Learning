Standard scientific Python imports

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Import datasets, classifiers and performance metrics


```python
from sklearn import datasets, metrics
from sklearn.metrics import confusion_matrix

from sklearn.neighbors import KernelDensity
from sklearn.neighbors import KNeighborsClassifier
from sklearn import tree
from sklearn.ensemble import GradientBoostingClassifier

from sklearn.model_selection import train_test_split
```

The adult dataset

```python
data = pd.read_csv('adult.csv')
data.head()
```

```python
data.describe()
```

```python
from numpy import nan
df=data.replace("?",nan)
```

```python
null_values=df.isnull().sum()
null_values=pd.DataFrame(null_values,columns=['null'])
j=1
sum_tot=len(df)
null_values['percent']=null_values['null']/sum_tot
round(null_values*100,3).sort_values('percent',ascending=False)
```

#### fill nan number

```python
df['native-country'].fillna(df['native-country'].mode()[0], inplace=True)
df['occupation'].fillna(df['occupation'].mode()[0], inplace=True)
df['workclass'].fillna(df['workclass'].mode()[0], inplace=True)
```

```python
null_values=df.isnull().sum()
null_values=pd.DataFrame(null_values,columns=['null'])
j=1
sum_tot=len(df)
null_values['percent']=null_values['null']/sum_tot
round(null_values*100,3).sort_values('percent',ascending=False)
```

#### Data processing

#### educational-num seems to be not so important

```python
df=df.drop(["educational-num"],axis=1)
```

#### also fnlwgt is not so important

```python
df=df.drop(["fnlwgt"],axis=1)
```

creat label

```python
df.income = [1 if each=='>50K' else 0 for each in data.income]
y = df.income
y.head()
```

As you can see i got labels as y and encoded them if bigger than 50K to 1 else 0.

```python
df.head()
```

```python
from sklearn import preprocessing
import pandas as pd
le = preprocessing.LabelEncoder()
```

```python
df[['age', 'workclass', 'education', 'marital-status', 'occupation',
       'relationship', 'race', 'gender', 'capital-gain', 'capital-loss',
       'hours-per-week', 'native-country']]=df[['age', 'workclass', 'education', 'marital-status', 'occupation',
       'relationship', 'race', 'gender', 'capital-gain', 'capital-loss',
       'hours-per-week', 'native-country']].apply(le.fit_transform)
```

```python
df.drop(['income'],axis=1).head()
```

Split data into train and test subsets

```python
from sklearn.model_selection import train_test_split
x_train , x_test, y_train, y_test = train_test_split(df,y,test_size=0.3,random_state=42)
```

## KNeighborsClassifier

Using adult data, Train a KNN Classifier and draw the confusion matrix according to the code.

```python
classifier = KNeighborsClassifier(n_neighbors=3).fit(x_train, y_train)
predicted = classifier.predict(x_test)
```

```python
print("Classification report for classifier %s:\n%s\n"
      % (classifier, metrics.classification_report(y_test, predicted)))
disp = confusion_matrix(y_test, predicted)
print("Confusion matrix:\n%s" % disp)
metrics.plot_confusion_matrix(classifier, x_test, y_test)
plt.show()
```

#### Train the classifier for different K,

```python
for k in range(1, 10):
    classifier = KNeighborsClassifier(n_neighbors=k).fit(x_train, y_train)
    predicted = classifier.predict(x_test)
    print("Classification report for classifier %s:\n%s\n"
          % (classifier, metrics.classification_report(y_test, predicted)))
    disp = confusion_matrix(y_test, predicted)
    print("Confusion matrix:\n%s" % disp)
```

## Parzen

#### now train a kernerDensity model with diffrent kernels,

```python
for i, kernel in enumerate(['gaussian', 'tophat', 'epanechnikov', 'exponential', 'linear', 'cosine']):
    classifier = KernelDensity(kernel=kernel, bandwidth=0.2).fit(x_train, y_train)
    
    log_dens = classifier.score_samples(x_test)
    plt.fill(x_test, np.exp(log_dens), c='green')
    plt.title("Best Kernel: " + classifier.kernel+" h="+"{:.2f}".format(classifier.bandwidth))
    plt.show()
```

## Decision Tree

#### train a decision tree model with find its accuracy

```python
classifier = tree.DecisionTreeClassifier().fit(x_train, y_train)
predicted = classifier.predict(x_test)

print("Classification report for classifier %s:\n%s\n"
      % (classifier, metrics.classification_report(y_test, predicted)))
disp = confusion_matrix(y_test, predicted)
print("Confusion matrix:\n%s" % disp)
```

#### Now change the model parameters to optimize it.

```python
classifier = tree.DecisionTreeClassifier(max_depth=10).fit(x_train, y_train)
predicted = classifier.predict(x_test)

print("Classification report for classifier %s:\n%s\n"
      % (classifier, metrics.classification_report(y_test, predicted)))
disp = confusion_matrix(y_test, predicted)
print("Confusion matrix:\n%s" % disp)
```

### Boosting

#### train a Boosting model for diffrrent max depth and different learning rate and find its accuracy for validation set 

```python
for max_depth in range(1,10):
    classifier = GradientBoostingClassifier(learning_rate=1, max_depth=max_depth, random_state=0).fit(x_train, y_train)
    predicted = classifier.predict(x_test)

    print("Classification report for classifier %s:\n%s\n"
          % (classifier, metrics.classification_report(y_test, predicted)))
    disp = confusion_matrix(y_test, predicted)
    print("Confusion matrix:\n%s" % disp)
```

```python
for lr in np.linspace(0.1,1,10):
    classifier = GradientBoostingClassifier(learning_rate=lr, max_depth=5, random_state=0).fit(x_train, y_train)
    predicted = classifier.predict(x_test)

    print("Classification report for classifier %s:\n%s\n"
          % (classifier, metrics.classification_report(y_test, predicted)))
    disp = confusion_matrix(y_test, predicted)
    print("Confusion matrix:\n%s" % disp)
```
